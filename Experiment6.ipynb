```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
upload = files.upload()
# Load the dataset
df = pd.read_csv('report.csv')

# Check column names to ensure they are correctly loaded
print(df.columns)

# Fill missing values for numerical columns using the median
df['Cl_thickness'].fillna(df['Cl_thickness'].median(), inplace=True)
df['Cell_size'].fillna(df['Cell_size'].median(), inplace=True)
df['Cell_shape'].fillna(df['Cell_shape'].median(), inplace=True)
df['Marg_adhesion'].fillna(df['Marg_adhesion'].median(), inplace=True)
df['Epith_c_size'].fillna(df['Epith_c_size'].median(), inplace=True)
df['Bare_nuclei'].fillna(df['Bare_nuclei'].median(), inplace=True)
df['Bl_cromatin'].fillna(df['Bl_cromatin'].median(), inplace=True)
df['Normal_nucleoli'].fillna(df['Normal_nucleoli'].median(), inplace=True)
df['Mitoses'].fillna(df['Mitoses'].median(), inplace=True)


# Detect outliers using IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Detect outliers in some columns
outliers_cl_thickness = detect_outliers_iqr(df, 'Cl_thickness')
outliers_cell_size = detect_outliers_iqr(df, 'Cell_size')

# Print number of outliers detected
print(f"Outliers in Cl_thickness: {len(outliers_cl_thickness)}")
print(f"Outliers in Cell_size: {len(outliers_cell_size)}")

# Remove outliers in 'Cl_thickness' if necessary
Q1 = df['Cl_thickness'].quantile(0.25)
Q3 = df['Cl_thickness'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers
df = df[(df['Cl_thickness'] >= lower_bound) & (df['Cl_thickness'] <= upper_bound)]

# Repeat the process for other columns if needed
```
